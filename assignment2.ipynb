**DATA TYPES**

**1.int (Integer)**] :
Represents whole numbers, positive or negative, without a decimal point.

**Example:** 5, -42, 0
Used for counting or performing arithmetic where decimals arenâ€™t needed.

**2.float (Floating-point)**:
Represents numbers with a decimal point or in exponential form.

**Example:** 3.14, -0.001, 2.5e3 (which is 2500.0)
Useful for precise calculations like measurements or scientific data.

**3.complex (Complex Number)**
Represents numbers with a real and imaginary part, written as a + bj where j is the imaginary unit.
**Example:** 3 + 4j, -2.5 + 0j
Common in engineering or advanced math scenarios.

**4.list (List):**
An ordered, mutable (changeable) collection of items, which can be of mixed types.

**Example:** [1, "hello", 3.14]
Great for storing and modifying sequences of data.

**5.tuple (Tuple):**
An ordered, immutable (unchangeable) collection of items, also allowing mixed types.

**Example**: (1, "hello", 3.14)
Used when you want a fixed, unalterable sequence.

**6.str (String):**
A sequence of characters, used to represent text.

**Example:** "hello", 'Python3', "" (empty string)
Essential for handling words, sentences, or any text data.

**7.set (Set):**
An unordered, mutable collection of unique items (no duplicates).

**Example:** {1, 2, 3}, {"apple", "banana"}
Handy for removing duplicates or performing math-like operations (e.g., unions, intersections).

**8.dict (Dictionary)**
An unordered, mutable collection of key-value pairs.

**Example:** {"name": "Alice", "age": 25}
Perfect for associating data, like a real-world dictionary or database lookup.

**9.bool (Boolean)**
Represents one of two values: True or False.

**Example:** True, False
Used for logical operations, conditionals, or binary decisions.

#Strings

```python
#upper(): Returns the string with all characters in uppercase.
s = 'hello'
print(s.upper())  # Output: 'HELLO'
#lower(): Returns the string with all characters in lowercase.
s = 'HELLO'
print(s.lower())  # Output: 'hello'
#strip(chars=None): Returns the string with leading and trailing characters removed (default is whitespace).


s = '  hello  '
print(s.strip())  # Output: 'hello'
#replace(old, new, count=-1): Returns a string with all occurrences of old replaced by new. The count parameter specifies the maximum number of replacements.


s = 'hello world'
print(s.replace('world', 'Python'))  # Output: 'hello Python'
#split(sep=None, maxsplit=-1): Returns a list of words in the string, separated by the delimiter sep. The maxsplit parameter specifies the maximum number of splits.


s = 'hello world'
print(s.split())  # Output: ['hello', 'world']
#join(iterable): Concatenates the strings in the iterable with the string as the delimiter.


s = ' '
print(s.join(['hello', 'world']))  # Output: 'hello world'
```

# Sets

```python
#add(item): Adds an item to the set.

st = {1, 2, 3}
st.add(4)
print(st)  # Output: {1, 2, 3, 4}
#remove(item): Removes an item from the set. Raises a KeyError if the item is not found.


st = {1, 2, 3}
st.remove(2)
print(st)  # Output: {1, 3}
#discard(item): Removes an item from the set if it is present.

st = {1, 2, 3}
st.discard(2)
print(st)  # Output: {1, 3}
#pop(): Removes and returns an arbitrary item from the set.

st = {1, 2, 3}
print(st.pop())  # Output: 1 (or any other element)
print(st)  # Output: {2, 3}
#clear(): Removes all items from the set.

st = {1, 2, 3}
st.clear()
print(st)  # Output: set()
#union(*others): Returns the union of the set with other sets.
st1 = {1, 2, 3}
st2 = {3, 4, 5}
print(st1.union(st2))  # Output: {1, 2, 3, 4, 5}
#intersection(*others): Returns the intersection of the set with other sets.

st1 = {1, 2, 3}
st2 = {2, 3, 4}
print(st1.intersection(st2))  # Output: {2, 3}
#difference(*others): Returns the difference of the set with other sets.

st1 = {1, 2, 3}
st2 = {2, 3, 4}
print(st1.difference(st2))  # Output: {1}
#symmetric_difference(other): Returns the symmetric difference of the set with another set.

st1 = {1, 2, 3}
st2 = {2, 3, 4}
print(st1.symmetric_difference(st2))  # Output: {1, 4}
```

# Dictionary

```python
#keys(): Returns a view object containing the dictionary's keys.


d = {'a': 1, 'b': 2}
print(d.keys())  # Output: dict_keys(['a', 'b'])
#values(): Returns a view object containing the dictionary's values.

d = {'a': 1, 'b': 2}
print(d.values())  # Output: dict_values([1, 2])
#items(): Returns a view object containing the dictionary's key-value pairs.


d = {'a': 1, 'b': 2}
print(d.items())  # Output: dict_items([('a', 1), ('b', 2)])
#get(key, default=None): Returns the value for the specified key if it exists, otherwise returns default.


d = {'a': 1, 'b': 2}
print(d.get('a'))  # Output: 1
print(d.get('c', 'Not Found'))  # Output: 'Not Found'
#pop(key, default=None): Removes the specified key and returns the corresponding value. If the key is not found, returns default.


d = {'a': 1, 'b': 2}
print(d.pop('a'))  # Output: 1
print(d)  # Output: {'b': 2}
#popitem(): Removes and returns an arbitrary key-value pair.


d = {'a': 1, 'b': 2}
print(d.popitem())  # Output: ('b', 2)
print(d)  # Output: {'a': 1}
#clear(): Removes all items from the dictionary.


d = {'a': 1, 'b': 2}
d.clear()
print(d)
#update([other]): Updates the dictionary with the key-value pairs from other, overwriting existing keys.

d = {'a': 1}
d.update({'b': 2, 'a': 3})
print(d)
```

# BOOL
